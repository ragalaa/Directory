```
data = f.get('GOOG, FB, MSFT, GS, C', start='2018-01-01', end='2019-07-01')
data2 = f.get('SPY', start='2018-01-01', end='2019-07-01')
data3 = f.get('GOOG', start='2018-01-01', end='2019-07-01')
print(data.head())
```

```

ax = data.rebase().plot(figsize=(12,5))
```
