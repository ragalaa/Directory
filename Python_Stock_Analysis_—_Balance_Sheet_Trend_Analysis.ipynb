```
import pandas as pd
import requests
import json
""" This was  best API for all financials info for free data to fethced
    https://financialmodelingprep.com/api
"""
bs = requests.get('https://financialmodelingprep.com/api/v3/financials/balance-sheet-statement/AAPL?period=Quarter')
bs = bs.json()
bsfinance = bs['financials']
bsfinal = pd.DataFrame.from_dict(bsfinance)
bsfinal = bsfinal.T
bsfinal.columns = bsfinal.iloc[0]
```

```
cols = bsfinal.columns
bsfinal = bsfinal.iloc[1:,]
bsfinal = bsfinal.iloc[:12,:4]
cols = bsfinal.columns
bsfinal[cols] = bsfinal[cols].apply(pd.to_numeric, errors='coerce')
bsfinal
```

```
assetsQ1 = bsfinal.iloc[11,0]
assetsQ2 = bsfinal.iloc[11,1]
assetsQ3 = bsfinal.iloc[11,2]
assetsQ4 = bsfinal.iloc[11,3]
allassets = [assetsQ1,assetsQ2,assetsQ3,assetsQ4]
bsfinal[cols] = (bsfinal[cols] / allassets)*100
pd.options.display.float_format = '{:.2f}%'.format
bsfinal
```
